```python
import time
```

```python
indonesian_negative_words = [
    'ngata', 'cemooh','olok',
    'jelek', 'menertawakan', 'cibiran',
    'nyinyir',
]

religious_words = [
    'Agama', 'Islam', 'Kristen',
    'Buddha', 'Hindu', 'Konghucu',
    'Katolik', 'Protestan',
]
```

```python
delay = 10  # delay in seconds
i = 1

# Loop over the list of strings and run the command for each string
for words in indonesian_negative_words:
    for religion in religious_words:
        query = words + ' ' + religion

        if i == 1:
            print(f"Scraping: {i}. {query}\n")
        else:
            print(f"\n\nScraping: {i}. {query}\n")

        i += 1

        # Construct the shell command with the above query
        # Resource 1 (most useful): https://stackoverflow.com/questions/57330956/ipython-notebook-using-quotes-and-dollar-to-pass-string-variable-as-argum#:~:text=Any%20input%20line,in%20your%20line.
        # Resource 2: (didn't work for me): https://stackoverflow.com/questions/61606054/passing-ipython-variables-as-string-arguments-to-shell-command
        !python scraper -t 1500 -q "$query" --latest

        # Wait for the specified delay
        time.sleep(delay)
```
